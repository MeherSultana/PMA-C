```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image
from IPython.display import display
```

```python

# Specify the path to the folder containing your images
folder_path = "Desktop/MANCS/2nd_semister/PM&C/Evan_photos"

# List all files in the folder
image_files = [file for file in os.listdir(folder_path) if file.endswith((".jpg"))]

# Create a 3 x 3 subplot grid
fig, axs = plt.subplots(3, 3, figsize=(10, 10))

# Loop through the image files and plot them in the subplots
for i, image_file in enumerate(image_files[15:24]):  # Display the first 9 images
    # Create the full file path
    image_path = os.path.join(folder_path, image_file)

    # Open the image
    img = Image.open(image_path)

    # Plot the image in the corresponding subplot
    row = i // 3
    col = i % 3
    axs[row, col].imshow(img)
    axs[row, col].set_title(f"Image {i + 1}")
    axs[row, col].axis('off')

# Adjust layout and display the subplots
plt.tight_layout()
plt.show()
```

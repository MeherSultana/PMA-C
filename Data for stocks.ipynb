```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Data for stock forcasting .csv'
```

```python
df = pd.read_csv(file_path)
```

```python
print("Head of the Data:")
print(df.head())
```

```python
x = df['Month']
y = df['Passengers']

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Data')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Data Plot')
plt.legend()
plt.grid(True)
```
